```python
#Checking the data in country code file
df_country = pd.read_excel(r'C:\Users\ragul\OneDrive\Desktop\Data_Analyst\Python\Pandas\Datasets\Zomatodataset\Country-Code.xlsx')
df_country
```

```python
# merging both datasets
final_df = pd.merge(df,df_country, on = 'Country Code',how = 'left' )
final_df.head()
```

```python
# No_of restaurents in each country
country_val = final_df.Country.value_counts().values
country_val
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
# Top 3 countries uses Zomato the most

plt.pie(country_val[:3],labels = country_names[:3],autopct='%1.2f%%');
```

Observation:
    1) Zomato maximum businnes happening in india(94%) after that USA(4.7%) then UK(0.8%)


```python
# Findout a relationship between rating
rating=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

    1) For 2148 there is no rating
    2) Whenever the rating is between 4.5 to 4.9 --> Excellent
    3) Whenever the rating is between 4.0 to 4.4 --> Very good
    4) Whenever the rating is between 3.5 to 3.9 --> good
    5) Whenever the rating is between 2.5 to 3.4 --> Average
    6) Whenever the rating is less then 2.4 --> poor

```python
plt.figure(figsize=(12,12))
ax = sns.barplot(x ='Aggregate rating',y='Rating Count',data = rating,hue='Rating color',palette = ['white','red','orange','yellow','lightgreen','green'],edgecolor='black',linewidth=2 );

for p in ax.patches:
    ax.annotate(f'{p.get_height():.0f}',  # The height of the bar (the value)
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Positioning
                ha='center', va='center',  # Align the text
                xytext=(0, 10),  # Offset the text a little above the bar
                textcoords='offset points')
```
