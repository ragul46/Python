# Conditional selection

```python
dict1 ={'Name':['Priyang','Aadhya','Krisha','Vedant','Parshv',
                'Mittal','Archana'],
                'Marks':[98,89,99,87,90,83,82],
                'Grades':['AA','AB','AA','AB','AC','BA','BB']
               }
df1=pd.DataFrame(dict1,index=['stu1','stu2','stu3','stu4','stu5',
                             'stu6','stu7'])
```

```python
# Students having marks greater then 90

X = df1['Marks'] > 90
df1[X]
```

```python
# Students having marks greater then 90
df1[df1['Marks'] > 90]
```

```python
# Students having AB grade

df1[df1['Grades']=='AB']
```

```python
# Students having marks greater then 80 and AB grade

df1[(df1['Grades']=='AB') & (df1['Marks']>80)]
```

```python
# Names of the student having marks greater then 90

df1[df1['Marks'] > 90]['Name']
```

```python
# Names of the student having AB grade
df1[df1['Grades']=='AB']['Name']
```

```python
#use loc function to access this
# Students having marks greater then 80 and AB grade

df1.loc[(df1['Marks'] > 80) & (df1['Grades'] == 'AB')]
```

```python
#use iloc function to access this
# Students having marks greater then 80 and AB grade
df1[(df1.iloc[:,1] > 80) & (df1.iloc[:,2]=='AB')]
```
