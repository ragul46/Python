```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_data = np.random.random(50) * 100
y_data = np.random.random(50) * 100
```

```python
#scatterplot
plt.scatter(x_data,y_data,c='red',s=50,marker='+',alpha=0.3)
```

```python
#lineplot
year = [i for i in range(2004,2020)]
weight = [90,93,100,98,89,90,96,94,93,92,94,88,77,79,98,97]
```

```python
plt.plot(year,weight,c='black',marker='.',markersize=5,lw=0.5)
plt.title('Weight vs Year')
plt.xlabel('Year')
plt.ylabel('Weight')
plt.grid(True)
```

```python
#More formating fucntion in charts
year = [2016,2017,2018,2019,2020]
income = [78,82,89,100,94]
income_ticks = [75,80,85,90,100]
plt.plot(year,income);
plt.title('Income of Bob',fontsize=20)
plt.xlabel('year')
plt.ylabel('Income')
plt.yticks(income_ticks,[f"${i}k"for i in income_ticks]);
```

```python
#Barplot
x = ['C++','python','JS','Java','go','R','C','Sql']
y = [20,30,23,23,43,32,32,23]
colors = ['red', 'green', 'red', 'green', 'red', 'green', 'red', 'green']
plt.bar(x, y, width=0.5, color=colors,edgecolor='black',align='edge')
plt.title('Online voting for favourite subject')
plt.xlabel('Subject')
plt.ylabel('No of Votes')
```

```python
#histogram
ages = np.random.normal(20,1.5,1000)
```

```python
plt.hist(ages,bins=10,alpha=0.7)
plt.grid(True)
```

```python
#Cumative histogram -> it's like running sum
plt.hist(ages,bins=10,alpha=0.7,cumulative=True)
plt.grid(True)
```

```python
#pie chart
languages = ['C++','python','JS','Java','go']
votes = [20,100,23,23,43]
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0']
explode = (0.1,0,0,0,0)
plt.pie(votes,labels = languages,frame=False,colors=colors,explode=explode,autopct='%.2f%%',startangle = 90);
#If you keep start angle as 90 then it will start from left to right(anti clock wise)
```

```python
#Boxplot
data = np.random.normal(loc=20, scale=5, size=100)  # 100 data points with mean 20 and std dev 5

# Creating the boxplot
plt.boxplot(data)

# Adding title and labels
plt.title('Boxplot of Random Data')
plt.ylabel('Values')

# Show the plot
plt.show()
```

```python
first = np.linspace(0,10,25)
second = np.linspace(10,25,25)
third = np.linspace(25,50,25)
fourth = np.linspace(50,75,25)
```

```python
data =np.concatenate((first,second,third,fourth))
plt.boxplot(data);
plt.grid(True)
```

```python
#Legend
stock_a = [100,23,43,45,23,21]
stock_b = [10,23,54,60,34,21]
stock_c = [60,63,54,25,60,70]
plt.plot(stock_a,label='google');
plt.plot(stock_b,label='Amazon');
plt.plot(stock_c,label='Flipkart');
plt.legend();
```

```python
stock_a = [100,23,43,45,23,91]
x = [2000,2001,2002,2003,2004,2005]
plt.plot(x,stock_a,label='google');
plt.legend();
```

```python
#Multiple figures in signle chart,

x = np.arange(1,101)

fig,axs = plt.subplots(2,2)

axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title('Sine wave')

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title('cos wave')

axs[1,0].plot(x,np.random.random(100))
axs[1,0].set_title('random function')

axs[1,1].plot(x,np.log(x))
axs[1,1].set_title('log function')
plt.tight_layout()
plt.show;
plt.savefig("fourplots.png",dpi=300,transparent=False,bbox_inches='tight')
```

```python
# 3D projection
ax=plt.axes(projection='3d')
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)
ax.scatter(x,y,z)
ax.set_title("3D projection")
```

```python
ax = plt.axes(projection='3d')
X=np.arange(-5,5,0.1)
Y=np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)
ax.plot_surface(X,Y,Z,cmap='Spectral');
```
